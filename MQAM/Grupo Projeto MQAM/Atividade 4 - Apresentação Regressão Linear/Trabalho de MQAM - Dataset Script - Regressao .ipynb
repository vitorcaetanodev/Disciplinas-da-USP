```python
import pandas as pd
import numpy as np

dataset = pd.DataFrame(pd.read_csv("/content/diet_data.csv", sep=',')).dropna()
```

```python
dataset
```

```python
from scipy import stats
shapiro_test = stats.shapiro(dataset['weight'])
shapiro_test
```

```python
rng = np.random.default_rng()
x = stats.norm.rvs(loc=5, scale=3, size=100, random_state=rng)
shapiro_test = stats.shapiro(x)
print(format(shapiro_test.pvalue, '.8f'))
```
